```python
def prepare_dataset(train_data, test_data, batch_size=64, device="cuda"):
    """Create torch dataloader from the numpy data.

    Parameters
    ----------
    train_data : numpy.ndarray
        The training data.
    test_data : numpy.ndarray
        The testing data.
    batch_size : int, optional
        The batch size for the DataLoader (default is 64).
    device : str, optional
        The device on which the data will be moved (default is cuda).

    Returns
    -------
    train_loader : torch.utils.data.DataLoader
        DataLoader object for the training data with tensors of shape (1,image_dim,image_dim).
    test_loader : torch.utils.data.DataLoader
        DataLoader object for the testing data with tensors of shape (1,image_dim,image_dim).
    """
    X = train_data.reshape(-1, 1, image_dim, image_dim)
    X = torch.tensor(X).float().to(device)
    train_loader = DataLoader(TensorDataset(X), batch_size=64, shuffle=True)

    X_test = test_data.reshape(-1, 1, image_dim, image_dim)
    X_test = torch.tensor(X_test).float().to(device)
    test_loader = DataLoader(
        TensorDataset(X_test), batch_size=batch_size, shuffle=False
    )

    return train_loader, test_loader
```

```python
def prepare_classification_dataset(
    train_data_x,
    train_data_y,
    test_data_x,
    test_data_y,
    batch_size=64,
    prob=False,
):
    """Create torch dataloader from the numpy data fro classification problem.

    Parameters
    ----------
    train_data_x : numpy.ndarray
        The training data x.
    train_data_y : numpy.ndarray
        The training data y.
    test_data_x : numpy.ndarray
        The testing data x.
    test_data_y : numpy.ndarray
        The testing data y.
    batch_size : int, optional
        The batch size for the DataLoader (default is 64).
    prob: bool, optional
        The flag indicating, that y vector is the vector of probabilities (default is False).
    device : str, optional
        The device on which the data will be moved (default is cuda).

    Returns
    -------
    train_loader : torch.utils.data.DataLoader
        DataLoader object for the training data with x,y tensors of shape (1,image_dim,image_dim).
    test_loader : torch.utils.data.DataLoader
        DataLoader object for the testing data with x,y tensors of shape (1,image_dim,image_dim).
    """
    X = train_data_x
    X = torch.tensor(X).float().to("cuda")
    if prob:
        Y = torch.tensor(train_data_y).float().to("cuda")
    else:
        Y = torch.tensor(train_data_y).type(torch.LongTensor).to("cuda")
    train_loader = DataLoader(TensorDataset(X, Y), batch_size=64, shuffle=True)

    X_test = test_data_x
    X_test = torch.tensor(X_test).float().to("cuda")
    if prob:
        Y_test = torch.tensor(test_data_y).float().to("cuda")
    else:
        Y_test = torch.tensor(test_data_y).type(torch.LongTensor).to("cuda")
    test_loader = DataLoader(
        TensorDataset(X_test, Y_test), batch_size=batch_size, shuffle=False
    )

    return train_loader, test_loader
```

### Image generation
 - Below functions generate an image (or multiplie images in parallel) using a `sarNET` model. It can generate an image from scratch (starting with a black patch), or continue generating from a part of existing image. Image is generated patch by patch.

#### Generation with regressor
In regression, we predict the values of the output pixels. In each step, we need to ensure that the pixel values are within the range [0,1] (in case of standarized dataset). 

Plain generation patch by patch results in deterministic outputs. To obtain diverse images, we need to add some randomness to the generation process. We achieve this by adding a random vector $V_R$ from the $N(0,1)$ distribution to the image vector. The random vector has a size of $1x784$. We can parametrize this vector to control the degree of randomness:

$$noise = V_R \cdot \alpha$$
$$V_R \sim N(0, \alpha ^ 2)$$

At the begining of a generation we select the value of alpha based on the $\alpha_{max}$ parameter. 

$\alpha = \alpha_{max} * U[0,1]$

Adding even small noise to all the patches results in poor generation. To enhance this process, in each generation step, we randomly select previous patches (with probability $p$) to which we add the previously generated noise instead of taking all of them.

We add noise only to the input patches because we don't want noise in the output image. We keep the generated patches in a separate vector.

In the end, we can refine the image by feeding it once again into the network. The output image usually is sharper and has less noise. It's important to keep track the first generated patch and add it back after refinement because the refining method will most likely lose it (black out).

To stabilize generation process, we black out the pixels below a certain threshold - typically the pixels below a small constant ~0.09 are set to black.

Usually we set $p = 0.3$ and $\alpha=0.8$

#### Generation with classifier
In the classifier, we process the image similarly, patch by patch, but instead of modeling the numerical values of the pixels, we model the probability of each pixel belonging to a specific class.

The input can be in three formats:
- the image is standarized, and we interpret the intesities of pixels  as the probabilities of class $1$. 

- the images is first quantized into $N$ classes, e.g. classes: 0,1,2,3,4. 

- similar to the previous point, but additionaly we convert the quantized images to one hot encoded format.

In the generation process we create each patch based on the probabilities returned by the model. We sample the class from theses probabilities. To enhance diversity, we can divide the initial probabilities by a constant $t$ (temperature). If $t < 1$ the model will be more certain about tis predictions, if $t > 1$, the model will be less certain.

```python
def get_i_patch(patch_idx, patch_dim, img_size):
    """
    Get the pixel indices of a specific patch in the image.
    """
    row = patch_idx // (img_size // patch_dim)
    col = patch_idx % (img_size // patch_dim)
    pxl_idxs = [
        (row * patch_dim + p_y) * img_size + col * patch_dim + p_x
        for p_y in range(patch_dim)
        for p_x in range(patch_dim)
    ]
    return pxl_idxs


def init_image(patch_dim, image=None, pixels_seed=0):
    """
    Initialize an image by setting specific patches to zero.
    One can use a specific image as a seed to further generation.
    """
    patch_size = patch_dim**2
    patches = n_pixels // patch_size

    if image is None:
        start_image = np.zeros(n_pixels)
        indexes_of_patches = list(range(patches - 1))
    else:
        start_image = copy.deepcopy(image).reshape(n_pixels)

        assert pixels_seed % patch_size == 0
        indexes_of_patches = list(
            range((pixels_seed // patch_size - 1), patches - 1)
        )
        for i_p in indexes_of_patches:
            idxs = get_i_patch(i_p + 1, patch_dim, image_dim)
            start_image[idxs] = 0

    return start_image, indexes_of_patches


def plot_image(image, show_plot=True):
    """
    Plot an image.
    """
    plt.imshow(image.reshape(28, 28), cmap="gray")
    if show_plot:
        plt.show()
```

```python
def generate_regression_image(model, patch_dim, alpha=0, patch_rand=0.3, image=None, pixels_seed=0, th=0, plot=True, refine=False):
    """
    Generate an image using a regression model by processing patches sequentially.

    Parameters
    ----------
    model : nn.Module
        The regression model used for generating the image.
    patch_dim : int
        The dimension of the patches to process.
    alpha : float, optional
        The degree of noise to add for diversity. Default is 0.
    patch_rand : float, optional
        The probability of randomly selecting previous patches to add noise. Default is 0.3.
    image : numpy.ndarray, optional
        The initial image to be used as a base for generation. If None, a blank image is used. Default is None.
    pixels_seed : int, optional
        The starting seed for the pixels to be zeroed out. Default is 0.
    th : float, optional
        The threshold for pixel values. Default is 0.
    plot : bool, optional
        Whether to plot the generated image. Default is True.
    refine : bool, optional
        Whether to refine the generated image by feeding it back into the model. Default is False.

    Returns
    -------
    numpy.ndarray
        The generated image as a numpy array.
    """

    start_image, indexes_of_patches = init_image(patch_dim, image=image, pixels_seed=pixels_seed)

    # Adjust alpha for noise if specified
    if alpha > 0:
        alpha = np.random.rand(1) * alpha

    # Generate noise to add to the image
    noised_image = abs(np.random.randn(start_image.shape[0])) * alpha
    first_patch = None

    min_pixel, min_th = 0, th
    max_pixel, max_th = 1, 1

    for idx in indexes_of_patches:
        patch_idxs = get_i_patch(idx, patch_dim, image_dim)
        for r in range(1, idx + 1):
            if np.random.rand(1) < patch_rand:
                patch_idxs += get_i_patch(r, patch_dim, image_dim)

        # Keep a copy of the original patch
        orig_patch = copy.deepcopy(start_image[patch_idxs])

        # Add noise to the current patch
        start_image[patch_idxs] += noised_image[patch_idxs]

        # Threshold pixel values
        start_image[patch_idxs][start_image[patch_idxs] < min_th] = min_pixel
        start_image[patch_idxs][start_image[patch_idxs] > max_th] = max_pixel

        # Reshape and convert the image to a tensor for the model
        X = start_image.reshape(1,1,28,28)
        X = torch.tensor(X).float().to("cuda")

        Y = model(X)
        start_image[patch_idxs] = orig_patch

        Y = Y.to("cpu").detach().numpy().reshape(-1)
        patch_idxs = get_i_patch(idx+1, patch_dim, image_dim)

        Y = Y[patch_idxs]

        Y[Y < min_th] = min_pixel
        Y[Y > max_th] = max_pixel

        # Identify the first generated patch
        if sum(Y) > (min_pixel * patch_dim * patch_dim + 0.05) and first_patch == None:
            first_patch = patch_idxs

        start_image[patch_idxs] = Y

    start_image = start_image.reshape(28, 28)

    # Optionally refine the generated image
    if refine:
        tmp = model(torch.tensor(start_image).reshape(1,1,28,28).float().cuda())
        tmp = tmp.detach().cpu().numpy().reshape(-1)
        tmp[tmp > max_th] = max_pixel
        tmp[tmp < min_th] = min_pixel
        tmp[first_patch] = start_image.reshape(-1)[first_patch]
        start_image = tmp.reshape(28,28)

    # Optionally plot the generated image
    if plot:
        plot_image(start_image, show_plot=False)

    return start_image
```

```python
def generate_classification_image(model, patch_dim, temp=1, image=None, pixels_seed=0, one_hot=False, plot=True):
    """
    Generate an image using a classification model by processing patches sequentially.

    Parameters
    ----------
    model : nn.Module
        The classification model used for generating the image.
    patch_dim : int
        The dimension of the patches to process.
    temp : float, optional
        The temperature parameter to control the randomness of class predictions. Default is 1.
    image : numpy.ndarray, optional
        The initial image to be used as a base for generation. If None, a blank image is used. Default is None.
    pixels_seed : int, optional
        The starting seed for the pixels to be zeroed out. Default is 0.
    one_hot : bool, optional
        Whether to use one-hot encoding for the input image. Default is False.
    plot : bool, optional
        Whether to plot the generated image. Default is True.

    Returns
    -------
    numpy.ndarray
        The generated image as a numpy array.
    """

    start_image, indexes_of_patches = init_image(patch_dim, image=image, pixels_seed=pixels_seed)

    # Convert the image to one-hot encoding if specified
    if one_hot:
        start_image = start_image.astype(int)
        ploted_image = start_image.astype(int)
        start_image = F.one_hot(torch.from_numpy(start_image.reshape(1, image_dim,image_dim)), num_classes=CLASS_NUM).permute(0,3,1,2).numpy()
        start_image = start_image.reshape(1, -1, image_dim*image_dim)

    softmax = nn.Softmax(dim=0)

    for idx in indexes_of_patches:

        if one_hot:
            X = start_image
            X = torch.tensor(X).float().to("cuda")
            X = X.reshape(X.size(0), -1)
        else:
            X = start_image.reshape(1,1,28,28)
            X = torch.tensor(X).float().to("cuda")

        # Get the model's prediction and apply temperature scaling
        Y = model(X).reshape(-1, 28*28) / temp
        Y = softmax(Y.clone().detach())

        Y = Y.to("cpu").detach().numpy().reshape(-1, image_dim * image_dim).transpose()
        patch_idxs = get_i_patch(idx+1, patch_dim, image_dim)

        Y = Y[patch_idxs] 

        # Sample the class from the probabilities
        chosen = np.zeros(patch_dim * patch_dim)
        for i in range(patch_dim * patch_dim):
            chosen[i] = np.random.choice(CLASS_NUM, p=Y[i])

        Y = chosen

        # Update the start image with the predicted patch
        if one_hot:
            ploted_image[patch_idxs] = Y
            Y = F.one_hot(torch.from_numpy(Y.astype(int).reshape(1, patch_dim,patch_dim)), num_classes=CLASS_NUM).permute(0,3,1,2).numpy()
            Y = Y.reshape(1, -1, patch_dim*patch_dim)
            start_image[0, :, patch_idxs] = Y.reshape(-1, CLASS_NUM)
        else:
            start_image[patch_idxs] = Y

    # Optionally plot the generated image
    if plot:
        if one_hot:
            plot_image(ploted_image, show_plot=False)
        else:
            plot_image(start_image, show_plot=False)
    return start_image
```

### Losses and metrics

We calculate the Fréchet Inception Distance (FID) score between generated images and MNIST test images. We use `mnist_frechet_distance` function from the tensorflow_gan package. TF_GAN is widely used in projects and research at Google.

Unlike the traditional FID calculation, `mnist_frechet_distance` doesn't use the InceptionV3 model. Instead, it uses the TF-GAN MNIST classifier, which has ~99% accuracy. This approach allows for quicker calculations. The function expects the input pixels values to be in the range [-1,1].

We implemented here losses of the form: $L(x,y) = \frac{1}{n} \cdot \sum_1^n |x-y|^a$ with different powers $a$.

#### SSIM loss (only for regression)

SSIM - Structural Similarity Index Measure. It measures the similarity between two images based on structural information. It calculates three features from an image:
    - Luminance (average of all the pixels)
    - Contrast (standard deviation of all the pixels)
    - Structure (normalized image)

The value of this loss is between -1 and 1. The higher the score, the more similar are pictures.


We used here PyTorch MS-SSIM implementation of this loss, which allows for the fast gradient computations.

```python
def get_fid_score_mnist(generated_images, N):
    """
    Calculate the Fréchet Inception Distance (FID) score between generated images and MNIST test images.
    
    Parameters
    ----------
    generated_images : list, numpy.ndarray or torch.tensor
        The generated images to be evaluated. Should be a iterable list of images with shape (N, 28, 28).
    N : int
        The number of images to be used for calculating the FID score.

    Returns
    -------
    float
        The FID score between the generated images and the real MNIST test images.
    """

    # Select N real images from the standardized MNIST test data
    real_images = standarized_test_data.reshape(-1, 1, 28, 28)
    real_images = real_images[np.random.choice(real_images.shape[0], N)]

    # Scale images to [-1, 1] range -> this is the range of pixels the "mnist_frechet_distance" expects
    real_images = (
        real_images.reshape(-1, 1, 28, 28).transpose(0, 2, 3, 1) - 0.5
    ) * 2
    real_images = tf.convert_to_tensor(real_images, dtype=tf.float32)

    # Prepare the generated images
    # Scale images to [-1, 1] range -> this is the range of pixels the "mnist_frechet_distance" expects
    gen_x = (
        np.array(generated_images).reshape(-1, 1, 28, 28).transpose(0, 2, 3, 1)
        - 0.5
    ) * 2
    gen_x = tf.convert_to_tensor(gen_x, dtype=tf.float32)

    # Calculate the number of batches for the FID computation
    num_batches = int(np.ceil(N / 500))
    num_batches = max(1, num_batches)
    score = mnist_frechet_distance(real_images, gen_x, num_batches)

    return score
```

```python
def ssim_loss(outputs, targets):
    """
    Compute the Structural Similarity Index Measure (SSIM) loss.
    """
    y_pred = outputs.reshape(-1, 1, 28, 28)
    y = targets.reshape(-1, 1, 28, 28)

    loss = 1 - ssim(y_pred, y, data_range=1, size_average=True)
    return loss

class CustomPowLoss():
    """
    Custom loss function that raises the absolute difference between outputs and targets to a specified power.
    """
    def __init__(self, power):
        self.power = power
    def __call__(self, outputs, targets):
        loss = abs(outputs - targets) ** self.power
        return loss.mean()
    
class CombinedLoss():
    """
    Combined loss function that sums two loss functions with specified weights.
    """
    def __init__(self, loss1, loss2, a1=1, a2=1):
        self.loss1, self.loss2 = loss1, loss2
        self.a1, self.a2 = a1, a2
    def __call__(self, outputs, targets):
        return self.a1 * self.loss1(outputs, targets) + self.a2 * self.loss2(outputs, targets)
    
def compute_error(model, test_loader, loss, loss_to_monitor):
    loss_sum = 0
    monitored_loss = 0
    with torch.no_grad():
        for x in test_loader:
            x = x[0]
            y_pred = model(x)
            l = loss(y_pred, x)
            monitored_loss += loss_to_monitor(y_pred, x)
            loss_sum += l

    print(f"Overall loss: {loss_sum / len(test_loader)}, Monitored loss: {monitored_loss / len(test_loader)}")
    return loss_sum / len(test_loader)

def compute_classification_error(model, test_loader, loss, loss_to_monitor):
    loss_sum = 0
    monitored_loss = 0
    with torch.no_grad():
        for x,y in test_loader:
            y_pred = model(x)
            l = loss(y_pred, y)
            monitored_loss += loss_to_monitor(y_pred, y)
            loss_sum += l

    print(f"Overall loss: {loss_sum / len(test_loader)}, Monitored loss: {monitored_loss / len(test_loader)}")
    return loss_sum / len(test_loader)
```

# Architecture

### SemiDense
The core of the sarNET is a sparse linear layer. It is designed so that, except the first neuron, each pixel/patch on the input is connected to the neurons corresponding to the subsequent pixel/patches. Additionaly, one can add connections from all the output neurons to the input latent vector, but in this implementation we won't use the latent vectors.

The SemiDense layer takes a flattened image as an input and process it pixel by pixel  or patch by patch (depending on the configuration). The patches are subsequent squares of the image and do not overlap. The layer can be configured in to ways:

- the 0-th output patch depends on the 0-th input patch, and the i-th output patch (for i>0) depends on the 0,...,(i-1)th input patches (used in the first layer)
- the i-th output patch depends on the 0,...,i-th input patches (used in deeper layers)

The image below illustrates the second configuration. As shown, the last input patch is not connected to any neurons because we want to predict the last output patch based on the previous ones.

<img src="docs/SemiDense_graph_no_latent.png" alt="SemiDense Graph" width="500">

The design allows us to feed the entire falttened image into a network and obtain an output image of the same size. There is no need to "shift" the sequence (forced teacher method) as in transformer models.

To preserve the autoregressive nature of the layer, the number of output pixels/patches should be the same as the number of input pixels/patches. The size of the input patches can vary the size of the output patches. For example, we can compress the image from patches of size 4x4 to patches of size 2x2 or 1x1. Conversely, we can expand the size of the patch to, for example, 8x8. Additionaly, we can add channels to the input and output image - the architecture remains the same, but instead of looking at/predicting the patch in a single channel, we look at/ predicit the patch in every channel.



### sarNET regressor

- The input image is fed into a stack of SemiDense layers. Within the stack of the SemiDense layers, the first layer is of type 1 - it means that the 0-th output patch depends on the 0-th input patch, and the i-th output patch (for i>0) depends on the 0,...,(i-1)th input patches. Subsequent layers are of type 2 - the i-th output patch depends on the 0,...,i-th input patches (used in deeper layers). We decided to use 3 SemiDense layers. The first one compresses the input patches, second one generates the image, and the last one refines the output.

- We experimented with different preprocessing techniques for the images, such as standardization, normalization, and quantization.

<img src="docs/sarNET_VAE_regressor.png" alt="SemiDense Graph" width="500">

- We also experimented with different types of loss functions
    - L1 loss
    - MSE loss
    - losses of the form: $L(x,y) = \frac{1}{n} \cdot \sum_1^n |x-y|^a$ with different powers $a$
    - SSIM loss

### sarNET classifier

- It has a vary similar design to the regressor, but it returns the probabilities of a given class for each pixel.

- Pixel classes can be interpreted in two ways:
    - a pixel can be 0 or 1, and we interpret the probability as the pixel intesity, thus obtaining grayscale pixels
    - a pixel can take values from 0 to 255, or if quantized to n classes, values from 0 to n.

<img src="docs/sarNET_VAE_classifier.png" alt="SemiDense Graph" width="500">

- We also experimented with feeding the network with probabilities of given pixel (only in the case of 2 classes)



# Training Regression

We conducted several experiments to compare how different loss functions influence the training process and the model's capability to generate and reconstruct images. We found that:

- As expected, the higher the power in the loss function, the more blurred the generated images become.
- L1 loss appears to provide the best results.
- SSIM significantly improves the training process. It speeds up the training and enables further reduction of the loss.
- Results with SSIM combined with L1 or MSE are similar. The lowest MSE during training with SSIM is around 0.025, while without SSIM it is around 0.035.
- We found that with smaller patches, the loss can be reduced even further (MSE 0.019), but it is harder to generate images with small patches.
- Reconstruction works well with small patches.
- The generation and reconstruction look very similar for patch dimensions of 4 and 7, with slightly better results for patch dimension 4.

```python
patch_dim = 4
hidden_size = 2
custom_loss = CustomPowLoss(2) # mse
layers = nn.Sequential(
    SemiDense(image_dim, image_dim // 2, 1, hidden_size, patch_dim, patch_dim // 2, shift=True, latent_dim=0),
    nn.ReLU(),
    SemiDense(image_dim // 2, image_dim, hidden_size, 2, patch_dim // 2, patch_dim, shift=False, latent_dim=0),
    nn.ReLU(),
    SemiDense(image_dim, image_dim, 2, 1, patch_dim, patch_dim, shift=False, latent_dim=0),
)

model_autoregressive_1 = sarNET(layers, custom_loss, image_dim=image_dim)
model_autoregressive_1.to("cuda")
```

# Training Classification

We conducted several experiments with the classification model. It doesn't perform as well as the regression model. However, it is easier to generate more diverse images with this method because we can naturally sample from the returned probabilities. We found that the model performs better with larger patches (patch dimension 4). The model struggles to generate high-quality images as the number of classes increases, though it can still reconstruct images.

The model that uses probabilities as both input and output generates far better results than other classification models. Unfortunately, it's difficult to produce diverse images with it. While it can reconstruct images, the reconstructed parts are visibly blurred. We conclude that it's better to use regression models for this task.

### Classification 2 classes, input is the probability of given class

```python
softmax_ssim = nn.Softmax(dim=1)

def ssim_loss_ce(outputs, targets):
    y_pred = outputs.reshape(-1, 2, 28, 28)
    y_pred = softmax_ssim(y_pred)
    y_pred = y_pred[:, 1, :, :].view(-1, 1, 28, 28)
    y = targets.reshape(-1, 2, 28, 28)
    y = y[:, 1, :, :].view(-1, 1, 28, 28)

    loss = 1 - ssim(y_pred, y, data_range=1, size_average=True)
    return loss

def generate_mod_classification_image(model, patch_dim, alpha=0, patch_rand=0.3, image=None, pixels_seed=0, th=0, plot=True, refine=False):
    start_image, indexes_of_patches = init_image(patch_dim, image=image, pixels_seed=pixels_seed)
    softmax = nn.Softmax(dim=0)

    # Adjust alpha for noise if specified
    if alpha > 0:
        alpha = 0.005 + np.random.rand(1) * (alpha - 0.005)

    # Generate noise to add to the image
    noised_image = abs(np.random.randn(start_image.shape[0])) * alpha

    first_patch = None

    min_pixel, min_th = 0, th
    max_pixel, max_th = 1,1

    for idx in indexes_of_patches:
        patch_idxs = get_i_patch(idx, patch_dim, image_dim)
        for r in range(1, idx + 1):
            if np.random.rand(1) < patch_rand:
                patch_idxs += get_i_patch(r, patch_dim, image_dim)

        # Keep a copy of the original patch
        orig_patch = copy.deepcopy(start_image[patch_idxs])

        # Add noise to the current patch
        start_image[patch_idxs] += noised_image[patch_idxs]

        # Threshold pixel values
        start_image[patch_idxs][start_image[patch_idxs] < min_th] = min_pixel
        start_image[patch_idxs][start_image[patch_idxs] > max_th] = max_pixel

        X = start_image.reshape(1,1,28,28)
        X = torch.tensor(X).float().to("cuda")

        Y = model(X)
        Y = model(X).reshape(-1, 28 * 28)
        Y = softmax(Y.clone().detach())
        start_image[patch_idxs] = orig_patch

        Y = Y.to("cpu").detach().numpy().reshape(-1, image_dim * image_dim).transpose()
        patch_idxs = get_i_patch(idx+1, patch_dim, image_dim)

        Y = Y[patch_idxs]
        Y = Y[:, 1]

        if sum(Y) > (min_pixel * patch_dim * patch_dim + 0.05) and first_patch == None:
            first_patch = patch_idxs

        start_image[patch_idxs] = Y

    start_image = start_image.reshape(28, 28)

    if refine:
        tmp = model(torch.tensor(start_image).reshape(1,1,28,28).float().cuda())
        tmp = tmp.reshape(-1, 28 * 28)
        tmp = softmax(tmp.clone().detach())
        tmp = tmp.to("cpu").detach().numpy().reshape(-1, image_dim * image_dim).transpose()
        tmp = tmp[:, 1]
        tmp[first_patch] = start_image.reshape(-1)[first_patch]
        start_image = tmp.reshape(28,28)

    if plot:
        plot_image(start_image, show_plot=False)

    return start_image
```

# Patch importance (custom attention)

In the first SemiDense layer of our models, we can calculate the activations of neurons in specific patches to determine the importance of each patch at the input. We do this by taking the product of pixels in a patch and the weights connected to this patch. Then, we sum the absolute values and interpret the outcome as the activation of the given patch on the input. By iterating over each previous patch, we can rank the patches based on their importance.


```python
def get_patches_importance(image, importance_checker, patch_num, top_k):
    """
    Visualize the importance of patches in an image using the importance checker.

    Parameters
    ----------
    image : numpy.ndarray
        The input image as a numpy array.
    importance_checker : SemiDenseImportance
        An instance of the SemiDenseImportance class to check patch importance.
    patch_num : int
        The patch number for which to check importance.
    top_k : int
        The number of top important patches to visualize.

    Returns
    -------
    None
        This function displays a plot showing the importance of patches in the image.
    """

    image = copy.deepcopy(image)
    image_torch = (
        torch.tensor(image.reshape(1, 1, image_dim, image_dim)).float().to("cuda")
    )
    image = np.stack((image,) * 3, axis=-1)

    important_patches = importance_checker.check_importance(patch_num, top_k, image_torch)

    patch_idxs = get_i_patch(patch_num, patch_dim, image_dim)
    image.reshape(-1, 3)[patch_idxs] = 1

    # Prepare the colormap and values for the important patches
    cmap = plt.cm.cool
    values = []
    for i in important_patches:
        values.append(i[0])
    norm = plt.Normalize(min(values), max(values))
    values = norm(values)
    colors = cmap(values)[:, :3]

    # Color the important patches based on their importance
    for i in range(len(important_patches)):
        p_n = important_patches[i][1]
        patch_idxs = get_i_patch(p_n, patch_dim, image_dim)
        image.reshape(-1, 3)[patch_idxs] = colors[i]

    # Plot the image with highlighted patches
    #fig, ax = plt.subplots()
    cax = plt.imshow(image, cmap="cool")
    #cbar = plt.colorbar(cax, ax=ax, norm=norm)
    #cbar.set_label("Patch importance")
    #plt.title(f"Patch importance for selected {patch_num} patch - colored as white")
    #plt.show()
```
